```python
regions = {
    "Central": ["Illinois","Indiana","Iowa","Kentucky","Michigan","Minnesota","Missouri","Ohio","Wisconsin"],
    "Northern": ["Connecticut","Delaware","Maine","Maryland","Massachusetts","New Hampshire","New Jersey","New York",
                "Pennsylvania","Rhode Island","Vermont"],
    "Pacific": ["Alaska","Arizona","California","Hawaii","Nevada","Oregon","Utah","Washington"],
    "Southern": ["Alabama","Arkansas","Florida","Georgia","Louisiana","Mississippi","North Carolina","South Carolina",
                "Tennessee","Virginia","West Virginia"],
    "Western": ["Colorado","Idaho","Kansas","Montana","Nebraska","New Mexico","North Dakota","Oklahoma","South Dakota",
                "Texas","Wyoming"]
}
```

```python
def region(state):
    for key in regions.keys():
        if state in regions[key]:
            return key
```

```python
bad_drivers["Region"] = bad_drivers["State"].apply(region)
```

```python
bad_drivers["Car Insurance Premiums ($)"].mean()
```

```python
bad_drivers.pivot_table(index="Region",aggfunc='mean')
```

```python
def stackhist(x, y, **kws):
    grouped = pd.DataFrame.groupby(x, y)
    data = [d for _, d in grouped]
    labels = [l for l, _ in grouped]
    plt.hist(data, histtype="barstacked", label=labels)
```

```python
stackhist(bad_drivers["Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding"],bad_drivers.Region)
plt.legend()
```

```python
bad_drivers["Unsafe Driving Total"] = (bad_drivers["Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding"] + 
                                 bad_drivers["Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired"] +
                                 (100 - bad_drivers["Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted"]))
```
