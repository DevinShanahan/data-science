```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt('/Users/devin/ML/machine-learning-ex2/ex2/ex2data1.txt',delimiter=",")
X = data[:,:2]
y = data[:,2]
m = len(y)
m,n = np.shape(X)
```

### Part 1: Plotting

```python
def plot_data(X,y,labels=["",""]):  
    pos = np.argwhere(y == 1)[:,0]
    neg = np.argwhere(y == 0)[:,0]
    fig = plt.figure(figsize=(8,8))
    plt.scatter(X[pos][:,0],X[pos][:,1],marker='+',s=100,c='y')
    plt.scatter(X[neg][:,0],X[neg][:,1],marker='o',c='black')
    
    plt.xlabel(labels[0])
    plt.ylabel(labels[1])
    
```

```python
plot_data(X,y,labels=['Exam 1 Score','Exam 2 Score'])
plt.legend(['admitted','not admitted'])
```

### Part 2: Compute Cost and Gradient

```python
X = np.concatenate((np.ones((m,1)), X),axis=1)
initial_theta = np.zeros(3)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def cost_function(theta, X, y):
    m = len(y)
    h = sigmoid(X.dot(theta.T))
    J = np.sum(np.multiply(-y,np.log(h)) - np.multiply((1-y), np.log(1 - h))) / m
    grad = (1/m) * np.dot(np.subtract(h,y).T,X)
    return J, grad
```

```python
cost, grad = cost_function(initial_theta, X, y)
```

```python
print(cost)
print(grad)
```

```python
test_theta = np.array([-24, 0.2, 0.2])
```

```python
cost, grad = cost_function(test_theta, X, y)
```

```python
print(cost)
print(grad)
```

### Part 3: Optimizing

```python
import scipy.optimize as opt
```

```python
theta = opt.minimize(cost_function, initial_theta, method='BFGS', jac=True, args= (X,y),
               options={'disp': True})
theta = theta.x
print(theta)
```

```python
def plotDecisionBoundary(theta, X, y):
    plot_data(X[:,1:], y)

    if X.shape[1] <= 3:
        plot_x = np.array([min(X[:,1])-2,  max(X[:,1])+2])
        plot_y = (-1 /theta[2]) * (theta[1] * plot_x + theta[0])
        plt.plot(plot_x,plot_y)

    else:
        u = np.linspace(-1, 1.5, 50);
        v = np.linspace(-1, 1.5, 50);

        z = np.zeros((len(u), len(v)));
        
        for i in range(1,len(u)):
            for j in range(1,len(v)):
                z[i,j] = np.dot(map_features(u[i], v[j]),theta)
        z = z.T
        plt.contour(u,v,z)
```

```python
plotDecisionBoundary(theta,X,y)
```

```python
plot_data(X[:,1:],y,labels=['Exam 1 Score','Exam 2 Score'])
plt.plot([(-theta[0] - theta[2]*x1) / theta[1] for x1 in range(30,100)],[x1 for x1 in range(30,100)])
print(theta)
```

### Part 1: Regularized Logistic Regression

```python
data = np.loadtxt('/Users/devin/ML/machine-learning-ex2/ex2/ex2data2.txt',delimiter=",")
X = data[:,:2]
y = data[:,2]
m = len(y)
```

```python
plot_data(X,y, labels = ['Microchip Test 1','Microchip Test 2'])
plt.legend(['passed','failed'])
```

```python
def map_features(X1,X2):

    degrees = 6
    try:
        mapped = np.ones(X1.shape).reshape((len(X1),1))
    except TypeError:
        mapped = np.ones((1,1))
    for i in range(1,degrees+1):
        for j in range(i+1):
            try:
                mapped = np.hstack((mapped,(((X1**(i-j))*(X2**j)).reshape((len(X1),1)))))
            except TypeError:
                mapped = np.hstack((mapped,(((X1**(i-j))*(X2**j)).reshape((1,1)))))
    return mapped
```

```python
X = map_features(X[:,0],X[:,1])
```

```python
X[1,:]
```

```python
initial_theta = np.zeros(28)
lamda = 1
```

```python
def cost_function_reg(theta, X, y, lamda):
    m = len(y)
    h = sigmoid(X.dot(theta.T))
    J = np.sum(np.multiply(-y,np.log(h)) - np.multiply((1-y), np.log(1 - h))) / m + sum(lamda /(2*m) * theta[1:]**2)
    h = sigmoid(np.dot(X, theta.T))
    grad = (1/m) * np.dot(np.subtract(h,y).T,X)
    grad[1:] = grad[1:] + (lamda / m) * theta[1:]
    return J, grad
```

```python
J, grad = cost_function_reg(initial_theta, X, y, lamda)
print(J, grad[:5])
```

```python
test_theta = np.ones(28)
lamda = 10
```

```python
J, grad = cost_function_reg(test_theta, X, y, lamda)
print(J, grad[:5])
```

### Part 2: Regularization and Accuracies

```python
import scipy.optimize as opt
lamda = 1
options = {'maxiter': 400, 'disp':True}
theta  = opt.minimize(cost_function_reg, initial_theta, args= (X,y,lamda), method='BFGS', jac=True, options=options)
print(theta.x)
```

```python
plotDecisionBoundary(theta.x, X, y)
plt.title('lambda ={}'.format(lamda))
```

### Pandas and Scikit-learn Implementation
#### Linear

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import seaborn as sns

names = ['Exam 1 Score', 'Exam 2 Score', 'Admitted']
data = pd.read_csv('/Users/devin/ML/machine-learning-ex2/ex2/ex2data1.txt',delimiter=",",names=names)
X = data.iloc[:,:2]
y = data.iloc[:,2]
```

```python
model = LogisticRegression(solver='lbfgs')
model.fit(X,y)
theta = model.intercept_.tolist()
coef = model.coef_.reshape(-1).tolist()
theta.extend(coef)
print(theta)
```

```python
sns.lmplot('Exam 1 Score','Exam 2 Score',data,hue='Admitted', palette={1:'y',0:'black'},size=7, 
           markers=['o','x'],fit_reg=False)
plt.plot([(-theta[0] - theta[2]*x1) / theta[1] for x1 in range(30,100)],[x1 for x1 in range(30,100)])
```

```python
names = ['Microchip Test 1', 'Microchip Test 2', 'Pass']
data = pd.read_csv('/Users/devin/ML/machine-learning-ex2/ex2/ex2data2.txt',delimiter=",",names=names)
X = data.iloc[:,:2]
y = data.iloc[:,2]
```

```python
sns.lmplot('Microchip Test 1','Microchip Test 2',data,hue='Pass', palette={1:'y',0:'black'},size=7, 
           markers=['o','x'],fit_reg=False)
```

```python
def feature_map(X1,X2):
    degrees = 6
    new = []
    for i in range(1,degrees+1):
        for j in range(i+1):
            new_col = X1**(i-j) * X2**j
            new.append(new_col)
    new_df = pd.DataFrame(new)
    return new_df.T
    
```

```python
X = feature_map(X['Microchip Test 1'],X['Microchip Test 2'])
```

```python
X.shape
```

```python
model = LogisticRegression(solver='lbfgs')
model.fit(X,y)
theta = np.append(model.intercept_,model.coef_)
print(theta)
```

```python
u = np.linspace(-1,1.5,50)
v = np.linspace(-1,1.5,50)
z = np.zeros((len(u),len(v)))
for i in range(1,len(u)):
    for j in range(1,len(v)):
        z[i,j] = np.dot(np.concatenate((np.ones((1,1)),feature_map(u[i], v[j])),axis=1),theta)
z = z.T
```

```python
sns.lmplot('Microchip Test 1','Microchip Test 2',data,hue='Pass', palette={1:'y',0:'black'},size=7, 
           markers=['o','x'],fit_reg=False)

plt.contour(u,v,z)
```
